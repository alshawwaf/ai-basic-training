# Explore 03 — What's Inside a Dataset?

Every dataset you'll touch in this program looks roughly the same:
rows are samples, columns are features, plus a column of labels.

In this notebook you poke around the `digits` dataset until that
shape feels obvious. Run each cell, then **change the numbers** and
run them again. The empty cells at the end are for your own questions.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

digits = load_digits()
print(type(digits))
print('Fields:', list(digits.keys()))
```

## 1. Shape

Two questions to answer before you read the output:
1. How many samples do you think this dataset has?
2. How many features per sample?

Now run the cell.

```python
print('data    shape :', digits.data.shape)
print('images  shape :', digits.images.shape)
print('target  shape :', digits.target.shape)
print('classes        :', digits.target_names)
```

Notice `data` and `images` describe the same 1797 samples. `data` is
flat (1797, 64), `images` is reshaped (1797, 8, 8). Two views of the
same numbers.

## 2. One sample, two ways

Change `i` to any number from 0 to 1796 and re-run.

```python
i = 0  # change me

print(f'sample {i} label: {digits.target[i]}')
print()
print('flat (64 numbers):')
print(digits.data[i])
print()
print('reshaped (8x8):')
print(digits.images[i].astype(int))

fig, ax = plt.subplots(figsize=(2.5, 2.5))
ax.imshow(digits.images[i], cmap='gray_r')
ax.set_title(f'label: {digits.target[i]}')
ax.axis('off')
plt.show()
```

## 3. Wrap it in a DataFrame

DataFrames are how you'll explore real datasets. Notice that the
`pixel_*` columns are just regular numbers — pandas doesn't know or
care that they came from an image.

```python
df = pd.DataFrame(digits.data, columns=[f'pixel_{i}' for i in range(64)])
df['target'] = digits.target
df.head()
```

```python
df.describe().loc[['min', 'max', 'mean']].round(2)
```

Look at the `min` row. Several pixels are stuck at 0 across the entire
dataset — they're at the corners of the image where no digit is ever
drawn. We'll come back to that in `explore_04_useless_pixels.py`.

## 4. Class counts

How balanced are the classes? You'll need this every time you split a
dataset for training.

```python
counts = df['target'].value_counts().sort_index()
print(counts)
print()
print(f'imbalance ratio: {counts.max() / counts.min():.2f} : 1')

counts.plot(kind='bar', figsize=(6, 3), color='#06d6e0')
plt.xlabel('digit')
plt.ylabel('samples')
plt.title('Samples per class')
plt.show()
```
