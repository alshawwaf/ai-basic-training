# Explore 3 — What Shape Is This Dataset?

You've seen one digit. Now let's find out what the full dataset looks like.

Run each cell and answer the question **before** moving to the next one.

```python
from sklearn.datasets import load_digits
import pandas as pd
import numpy as np

digits = load_digits()
print("Dataset loaded.")
print(f"Type: {type(digits)}")
print(f"Fields: {list(digits.keys())}")
```

### Question 1
What fields does the dataset have? What do you think each one contains?

Try printing a few:

```python
# What's in digits.data?
print("digits.data shape:", digits.data.shape)
print("digits.target shape:", digits.target.shape)
print("digits.images shape:", digits.images.shape)
```

### Question 2
`digits.data` has shape `(1797, 64)`. What do the two numbers mean?

- 1797 = ______
- 64 = ______

```python
# Let's look at the first 5 samples as a table
df = pd.DataFrame(digits.data, columns=[f"pixel_{i}" for i in range(64)])
df["label"] = digits.target

# Show selected columns to keep it readable
df[["pixel_0", "pixel_10", "pixel_32", "pixel_63", "label"]].head()
```

### Question 3
Each row is one ______. Each column (except label) is one ______.

**Challenge:** Change the row index below to look at different samples. Try rows `0`, `100`, `500`, `1700`.

What happens if you try row `1797`?

```python
# Change this number and re-run the cell
ROW = 0

print(f"Sample {ROW}:")
print(f"  Label: {digits.target[ROW]}")
print(f"  First 10 pixel values: {digits.data[ROW, :10]}")
print(f"  As an image:")

import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(figsize=(3, 3))
ax.imshow(digits.images[ROW], cmap="gray_r", interpolation="nearest")
ax.set_title(f"Sample {ROW} — label: {digits.target[ROW]}", fontsize=12)
ax.axis("off")
plt.show()
```

### Question 4
How many different digits are in the dataset? Are they evenly distributed?

**Predict first:** Do you think all 10 digits have roughly equal numbers of samples, or are some more common?

```python
print("Unique labels:", np.unique(digits.target))
print(f"Number of classes: {len(np.unique(digits.target))}")
print()
print("Samples per class:")
for d in range(10):
    count = (digits.target == d).sum()
    print(f"  Digit {d}: {count} samples")
```

### What you discovered

- The dataset has **1,797 samples** (images) and **64 features** (pixels per image)
- Each pixel is a number from **0** (white) to **16** (black)
- There are **10 classes** (digits 0-9), roughly evenly distributed
- Some pixels (like corners) **never change** — they carry zero information
